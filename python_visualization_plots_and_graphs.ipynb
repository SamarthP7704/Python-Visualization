```python
import pandas as pd
import seaborn as sns   # Why sns?  It's a reference to The West Wing
import matplotlib.pyplot as plt  # seaborn is based on matplotlib
sns.set(color_codes=True) # adds a nice background to the graphs
%matplotlib inline
# tells python to actually display the graphs
```

```python
auto = pd.read_csv('sample_data/california_housing_test.csv')
```

```python
auto.head()
```

```python
sns.distplot(auto['population']);
```

```python
# we can turn the kde off and put a tic mark along the x-axis for every data point with rug
sns.distplot(auto['population'], kde=False, rug=True);
```

```python
sns.jointplot(auto['population'],auto['households']);
```

```python
sns.jointplot(auto['population'],auto['households'],kind="hex");
```

```python
sns.jointplot(auto['population'],auto['households'],kind="kde");
```

```python
sns.pairplot(auto[['latitude','housing_median_age','total_rooms']]);
```

```python
sns.stripplot(auto['latitude'], auto['total_rooms'], jitter=True);
```

```python
sns.swarmplot(auto['latitude'], auto['total_rooms']);
```

```python
sns.boxplot(auto['population'], auto['median_income']);
```
